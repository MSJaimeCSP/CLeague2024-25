```python
def informe_calidad(df: pd.DataFrame) -> pd.DataFrame:
    """Devuelve un DataFrame resumen con los parámetros de calidad por columna."""
    resumen = pd.DataFrame({
        "tipo":       df.dtypes.astype(str),
        "no_nulos":   df.notna().sum(), 
        "nulos":      df.isna().sum(),  # equivalente a df.isnull().sum()
        "pct_nulos":  (df.isna().sum() / len(df) * 100).round(2),
        "unicos":     df.nunique(),
        "pct_unicos": (df.nunique() / len(df) * 100).round(2),
    })
    return resumen

informe_calidad(df)
```

# Limpieza de nombres

```python
import re

sospechosos = df["player"].drop_duplicates()

for nombre in sospechosos:
    if re.search(r"[a-z][A-Z]", nombre):
        print(nombre)

df["player"] = df["player"].str.replace(
    r'(?<=[a-z])(?=[A-Z])',
    ' ',
    regex=True
)

df["player"].drop_duplicates().sort_values().to_csv(
    "jugadores_unicos.csv",
    index=False
)
```

```python
# Eliminar espacios al principio y final
df["player"] = df["player"].str.strip()

# Separar palabras pegadas por mayúsculas (incluyendo acentos)
df["player"] = df["player"].str.replace(
    r'(?<=[a-záéíóúñü])(?=[A-ZÁÉÍÓÚÑÜ])',
    ' ',
    regex=True
)


sospechosos = [
    n for n in df["player"].drop_duplicates()
    if re.search(r'(?<=[a-záéíóúñü])(?=[A-ZÁÉÍÓÚÑÜ])', n)
]

print(sospechosos)

df[df["player"].str.contains(r"\?", na=False)]["player"].unique()

raros = sorted(
    df[df["player"].str.contains(r"\?", na=False)]["player"].unique()
)

for r in raros:
    print(r)

pd.Series(raros).to_csv(
    "nombres_raros.csv",
    index=False,
    encoding="utf-8-sig"
)
```
